```python
import csv
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
header = ['Name', 'Age', 'Height', 'Gpa']

with open('Student_data.csv', 'w', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    num = 1
    for i in range(1000):
        first_names=('John','Andy','Joe', 'ALi', 'Umar', 'maryam', 'Ayesha', 'usman')
        last_names=num
        Name =" ".join(random.choice(first_names)+" "+str(last_names))
        num = num + 1
        gpa = random.uniform(2.5, 4)
        gpa = round(gpa, 2)
        writer.writerow([Name, (random.randint(18, 24)), (random.randint(60, 75)), gpa])    
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import OneHotEncoder
import joblib
```

```python
data = pd.read_csv('Student_data.csv')
data['Gpa']=data['Gpa'].astype('category')
data['Gpa']=data['Gpa'].cat.codes
enc=OneHotEncoder()
enc_data=pd.DataFrame(enc.fit_transform(data[['Gpa']]).toarray())
stu_data=data.join(enc_data)
```

```python
# load the iris dataset as an example

stu_data = pd.read_csv('Student_data.csv')

# store the feature matrix (X) and response vector (y)
X = np.array(stu_data['Age'])
y = np.array(stu_data['Gpa'])

# splitting X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# training the model on training set
knn = KNeighborsClassifier(n_neighbors=3)
y_train = y_train.astype('int')
knn.fit(X_train.reshape(-1,1), y_train.reshape(-1,1))

# making predictions on the testing set
y_pred = knn.predict(X_test.reshape(-1, 1))

# comparing actual response values (y_test) with predicted response values (y_pred)
y_test = y_test.astype(int)

print("kNN model accuracy : ", metrics.accuracy_score(y_test, y_pred)*100)

# showing predictions
print("So knn Predicted values are : ", y_pred)

```
